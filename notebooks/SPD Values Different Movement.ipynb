```python
%matplotlib inline
import pandas as pd
import numpy as np
DATA_DIR = '../data'
```

```python
df = pd.read_excel(f'{DATA_DIR}/change_values_am_peak.xlsx', 'Sheet1')
```

```python
di_matrix = pd.read_excel(f'{DATA_DIR}/distance_matrix_am_peak.xlsx', 'Sheet1')
```

```python
year_wise = df.groupby("Year")
```

```python
def find_diff_label_spd(label_group):
    movements = list(label_group.groups.keys())
    for name, group in label_group:
        movements.remove(name)
        if not movements:
            pass
        else:
            for label in movements:
                di_matrix.ix[di_matrix.index.isin(group['Station']), list(label_group.get_group(label)['Station'].values)] += 1
```

```python
def find_spd_by_label(group):
    labels = group.groupby("Label")

    find_diff_label_spd(labels)
```

```python
for name, group in year_wise:
        year_group = group[['Station','% Change','Label']]
        find_spd_by_label(year_group)
```
